```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
data = pd.read_csv("IC_IF_IH/IH-60-Difference-2.txt")
data.head()
```

```python
data['volume_total']=data.volume_thousand*1000+data.volume
data.drop('volume_thousand',axis=1,inplace = True)
data.drop('volume',axis=1,inplace = True)
```

```python
data.rename(columns = {'volume_total':'volume'},inplace=True)
data
```

```python
data['Date'] = data['date'].map(lambda x:x.split(' ')[0])
data['time'] = data['date'].map(lambda x:x.split(' ')[1])
data.drop('date',axis=1,inplace = True)
data
```

```python
data = data[['Date','time','open','high','low','close','volume']]
data.rename(columns = {'Date':'date'},inplace=True)
data
```

```python
def transform_to_date(x):
    try:
        return datetime.strptime(str(x), "%Y/%m/%d")
    except Exception as e:
        print(e)
```

```python
data['date'] = data['date'].map(
    lambda x: transform_to_date(x)
)
```

```python
data
```

```python
type(data.time[0])
```

```python
def transform_to_timedelta(x):
    hour = int(x[0: 2])
    minute = int(x[3: 5])
    
    timedelta_object = timedelta(
        hours=hour,
        minutes=minute
    )
    return timedelta_object
```

```python
data.time = data.time.apply(
    lambda x: transform_to_timedelta(x)
)
```

```python
type(data.time[0])
```

```python
data
```

```python
def get_OHLC(data_under_given_date):
    open_price = data_under_given_date.iloc[0, :]['open']
    high_price = data_under_given_date.high.max()
    low_price = data_under_given_date.low.min()
    close_price = data_under_given_date.iloc[-1, ]['close']
    volume = data_under_given_date.volume.sum()
    
    return pd.DataFrame([[
        open_price, high_price, low_price, close_price, volume
    ]])
```

```python
data.groupby('date')
```

```python
OHLC = data.groupby('date').apply(lambda x: get_OHLC(x))
OHLC.tail()
```

```python
OHLC.reset_index(inplace=True)
OHLC.head()
```

```python
OHLC.drop(columns='level_1', inplace=True)
OHLC.head()
```

```python
OHLC.columns = [
    'Date', 'Open', 'High', 'Low', 'Close', 'Volume'
]
OHLC.head()
```

```python
OHLC.to_csv('IC_IF_IH/IH_daily.csv', index=False)
```
